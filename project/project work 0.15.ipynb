```python
#read data
# the encoding is different so I had to use heading 0 and unicode escapte encoding to read it.
df = pd.read_csv('WB_data.csv',delimiter = '\t', encoding = 'unicode_escape', dtype={'Country Name':str, 'Country Code':str, 'Series Name':str, 'Series Code':str})

#replace the ".."s with NaNs for easier workflow
#df = df.replace("..", np.nan)

#and remove unnecessary Country Code and Series Code columns
#df = df.drop(['Country Code', 'Series Code'],axis = 1)

#display the first 5 rows
df.head()
```

```python
#name the colmuns so we can change the data type of the columns using a loop
Col_Names = list(df.columns)

for Col in Col_Names[2:]:
    df[Col] = pd.to_numeric(df[Col], errors = 'coerce')
for Col in Col_Names[:2]:
    df[Col] = df[Col].astype('str')


df.dtypes
```
